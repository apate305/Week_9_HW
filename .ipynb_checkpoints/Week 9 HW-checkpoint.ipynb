The Monty hall problem is a probability illusion. The problem states that if you have 3 doors, one of which has a prize behind it, and you choose the door youâ€™d like to open and are then shown one of the other doors that does not have the prize, then you are better off switching your choice because it increases your probability of winning.

Here is how it works: We may think that by knowing that one of the 3 doors does not have the prize, we have a 50/50 chance of winning. But this is not the case. Hereâ€™s why: Before you make your choice, you have 1 in 3 chances of getting the correct door. You also have 2 out of 3 chances of getting the incorrect door. So you have more of a chance of losing than you do winning. After you make your choice, the host will show you whatâ€™s behind one  of the two remaining doorsâ€”the two remaining doors will most likely be a correct door and an incorrect door since you had more of a chance of choosing the incorrect door to begin with. We also know that the host will never reveal the correct door. So out of the two remaining doors, he will chose to show you the incorrect one. That means the other door will most likely be correct. This is why by switching your choice, you are increasing your probability of winning. 

To be exact, you are increasing your probability by 66% compared to 33% which is the probability that you start off with. To prove this, you can make a table of all the possible outcomes. If you add up the amount of times you do not swtich, you only have 33% chance of winning. But if you add up the amount of times you do switch you ahve a 66% chance of winning.
